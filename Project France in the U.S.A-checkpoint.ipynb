## Exploring the area around The Amazon HQ



```python
radius = 1500
LIMIT = 100
url2 = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
    CLIENT_ID, 
    CLIENT_SECRET, 
    VERSION, 
    Amazon_Lat, 
    Amazon_Long, 
    radius, 
    LIMIT)
url2
```

```python
results = requests.get(url2).json()
```

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
venues = results['response']['groups'][0]['items']
    
nearby_venues = pd.json_normalize(venues) 

filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]

nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues.head(10)
```

```python
print('There are {} uniques categories.'.format(len(nearby_venues['categories'].unique())))
```

```python
amazon_nv_df = nearby_venues['categories'].value_counts()
amazon_nv_df.head()
```

```python
# one hot encoding
Amazon_onehot = pd.get_dummies(nearby_venues[['categories']], prefix="", prefix_sep="")

Amazon_onehot['categories'] = nearby_venues['categories'] 

fixed_columns = [Amazon_onehot.columns[-1]] + list(Amazon_onehot.columns[:-1])
Amazon_onehot = Amazon_onehot[fixed_columns]
```

```python
Amazon_grouped = Amazon_onehot.groupby('categories').sum().reset_index()
Amazon_grouped
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

columns = ['categories']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

Amazon_venues_sorted = pd.DataFrame(columns=columns)
Amazon_venues_sorted['categories'] = Amazon_grouped['categories']

for ind in np.arange(Amazon_grouped.shape[0]):
    Amazon_venues_sorted.iloc[ind, 1:] = return_most_common_venues(Amazon_grouped.iloc[ind, :], num_top_venues)

Amazon_venues_sorted
```
